```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. Pobranie danych
df = yf.download('NTDOY', period='max')
#.. yfinance zwraca dataframe w 2D, a potrzebujemy series w 1D.  Uyjemy .squeeze()
close_price = df['Close'].squeeze() 


fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(df.index, close_price, color='#6588b8', linewidth=2, label='Closing price')
ax.fill_between(df.index, close_price, color='#6588b8', alpha=0.1)

ax.set_title('History of Closing Price of Nintendo (NTDOY)', fontsize=24, pad=20)
ax.set_ylabel('Price in USD', fontsize=16)
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend(fontsize=12, loc='upper left')
ax.set_xlim(df.index.min(), df.index.max()) # Ustawienie limitw X na zakres danych
ax.set_ylim(0,30)

pogo_date = pd.Timestamp('2016-07-06')
price_at_date = close_price.asof(pogo_date)
ax.annotate(
    'Pokmon GO\n(6.07.2016)', # Tekst (uywamy \n dla nowej linii)
    xy=(pogo_date, price_at_date),          # Dokd strzaka celuje (Data, Cena)
    xytext=(-40, 100),                      # Gdzie jest tekst (przesunicie w punktach: w lewo, w gr)
    textcoords='offset points',             # Interpretacja xytext jako przesunicia
    arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle="arc3,rad=.2"), # Styl strzaki
    fontsize=11,
    ha='center', # Wyrwnanie tekstu w poziomie
    # Dodajemy biae to pod tekstem, eby by czytelny na siatce
    bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="gray", alpha=0.9)
)

img = plt.imread('images/images/61.png')
imagebox = OffsetImage(img, zoom=0.25) 
ab = AnnotationBbox(imagebox, xy=(pd.Timestamp('2024-01-01'), 2.45), frameon=False) #na minusie bo mamy odwrcone osie
ax.add_artist(ab) 

plt.tight_layout()
plt.show()
```
