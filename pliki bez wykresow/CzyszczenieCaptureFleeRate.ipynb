```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df2=pd.read_csv('pokemon_capture_flee.csv', sep=';')
df2=df2.dropna(subset='#').reset_index(drop=True)
df2=df2.drop_duplicates(subset='#', keep='first').reset_index(drop=True)
df2=df2.drop_duplicates(subset='Pokémon', keep='first').reset_index(drop=True)
df2['Base capture rate'] = (df2['Base capture rate'].str.replace('%', '', regex=False).astype(float)/100)
df2['Base flee rate'] = df2['Base flee rate'].str.replace('-', '0%') #myślnik oznacza brak zdefiniowanej wartości ucieczki - czyli ma 0% na ucieczkę
df2['Base flee rate'] = (df2['Base flee rate'].str.replace('%', '', regex=False).astype(float)/ 100)
df2.drop([904], axis=0, inplace=True) #axis=0 znaczy wiersz
df2
```

```python
df2.to_csv('capture_flee_rate')
```
